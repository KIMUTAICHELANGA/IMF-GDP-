### Time Series Plot for Global GDP Trends:

```python
import pandas as pd
import plotly.express as px

# Aggregating GDP by year
years = [str(year) for year in range(1980, 2029)]
global_gdp = data[years].sum()

# Creating a DataFrame for the plot
gdp_trends = pd.DataFrame({
    'Year': years,
    'Global GDP': global_gdp
})

# Creating the interactive line plot with customizations
fig = px.line(gdp_trends, x='Year', y='Global GDP', title='Global GDP Trends (1980-2028)')

# Customizing the layout
fig.update_layout({
    'plot_bgcolor': 'rgba(0, 0, 0, 0)',  # Transparent background inside the plot
    'paper_bgcolor': 'rgba(255, 255, 255, 1)',  # Background color outside the plot
    'title_font': {'size': 24},
    'title_x': 0.5,  # Center the title
    'xaxis_title': 'Year',
    'yaxis_title': 'Global GDP (in Trillions)',
    'xaxis': {
        'showline': True, 'showgrid': False, 'showticklabels': True,
        'linecolor': 'rgb(204, 204, 204)', 'linewidth': 2,
        'ticks': 'outside', 'tickfont': {'family': 'Arial', 'size': 12, 'color': 'rgb(82, 82, 82)'},
    },
    'yaxis': {
        'showgrid': True, 'zeroline': False, 'showline': False, 'showticklabels': True,
    },
    'autosize': True,
    'margin': {'autoexpand': True},
    'showlegend': False  # Hide the legend if it's not necessary
})

# Setting line and marker colors
fig.update_traces({
    'line': {'color': 'RoyalBlue'},
    'marker': {'color': 'RoyalBlue', 'size': 8}
})

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Calculating the year-on-year percentage change in GDP for each country
gdp_growth = data.set_index('Country').pct_change(axis='columns').reset_index()

# Melting the DataFrame from wide to long format to make it suitable for plotting with Plotly Express
long_format_growth = gdp_growth.melt(id_vars='Country', var_name='Year', value_name='GDP Growth Rate')

# Removing the first year since percentage change from the previous year cannot be calculated
long_format_growth = long_format_growth.dropna()

# Creating the interactive bar chart
fig = px.bar(long_format_growth, x='Year', y='GDP Growth Rate', color='Country', title='GDP Growth Rate Year-on-Year for Each Country')

# Customizing the layout
fig.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Transparent background inside the plot
    paper_bgcolor='rgba(255, 255, 255, 1)',  # Background color outside the plot
    title_font_size=22,
    xaxis=dict(
        title='Year',
        showline=True,
        showgrid=True,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)'
        ),
    ),
    yaxis=dict(
        title='GDP Growth Rate (%)',
        showgrid=True,
        zeroline=False,
        showline=False,
        showticklabels=True
    ),
)

# To improve readability, you might want to limit the number of countries shown or create separate plots for each country
# For a large number of countries, a line plot might be more appropriate

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Filtering out the countries of interest
countries_of_interest = ['Canada', 'Germany', 'Mexico', 'United States','']
filtered_data = data[data['Country'].isin(countries_of_interest)]

# Creating a DataFrame with necessary data
map_data = pd.DataFrame({
    'Country': filtered_data['Country'],
    'GDP': filtered_data['2028'],
    'ISO Alpha-3': ['CAN', 'DEU', 'MEX', 'USA']  # ISO Alpha-3 country codes
})

# Creating the interactive choropleth map
fig = px.choropleth(
    map_data,
    locations='ISO Alpha-3',
    color='GDP',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,
    title='GDP by Country for the Year 2028'
)

# Customizing the layout
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations=[dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Source: <a href="https://www.imf.org/">\
            International Monetary Fund</a>',
        showarrow=False
    )]
)

# Show the plot
fig.show()
```
